```python
#importing libraries
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import numpy as np
```

```python
url='https://www.imdb.com/list/ls081235533/'
```

```python
#requesting to allow scrapping
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')
```

```python
#request status
response
```

```python
# variables Declaration
movie_name=[]
year=[]
runtime=[]
rating=[]
metascore=[]
votes=[]
```

```python
movie_data=soup.findAll('div',attrs={'class':'lister-item mode-detail'})
```

```python
for store in movie_data:
    name=store.h3.a.text
    movie_name.append(name)
    
    release_year=store.h3.find('span',class_='lister-item-year text-muted unbold').text
    year.append(release_year)
    
    time=store.p.find('span',class_='runtime').text
    runtime.append(time)
    
    rate=store.find('span',class_='ipl-rating-star__rating').text
    rating.append(rate)

    meta=store.find('span',class_='metascore').text.replace(' ','') if store.find('span',class_='metascore') else 'no meta score given'
    metascore.append(meta)
    
    values=store.find_all('span',attrs={'name':'nv'})
    vote=values[0].text
    votes.append(vote)
    
    gross1=values[1].text if len(values)>1 else'^^'
    
        
```

```python
#list containing 100 movie names
movie_name
```

```python
#list containing 100 movie release dates
year
```

```python
movie_DF=pd.DataFrame({'movie name':movie_name,'year of release':year,'runtime':runtime,'movie rating':rating,'meta score':metascore,'votes':votes})
```
